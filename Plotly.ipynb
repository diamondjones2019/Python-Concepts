```python
import pandas as pd
import plotly.express as px
```

```python
#What day had the highest opened price 
zoom = pd.read_csv('Zoom.csv')

fig = px.line(zoom, x = 'Date', y = 'Open', title = 'What day had the highest opened price?')
fig.show()
```

```python
#What is Zoom highest traded price by date
import plotly.graph_objects as go
import pandas as pd

zoom = pd.read_csv('Zoom.csv')

fig = go.Figure()

fig.add_trace(go.Bar(x=zoom['Date'], 
                    y=zoom['High'],
                     name ="High",))

fig.show()
```

```python
#find the Zooms highest closed price by date
import plotly.graph_objects as go
import pandas as pd

zoom = pd.read_csv('Zoom.csv')

fig = go.Figure()

fig.add_trace(go.Bar(x=zoom['Date'], 
                    y=zoom['Close'],
                     name ="Close",))

fig.show()
```
